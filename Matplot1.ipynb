```python
#5.	Draw a Pie Chart showing the share of cases of top 4 states and cases of all remaining states as other. 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('covid.csv')
df.head
df1=df.iloc[0:4,:]
df2=df.loc[5:10]

print(df1)
print(df2)
plt.subplot(2,2,1)
plt.pie('May-20',labels='State',autopct='%1.1f%%', data=df1)
plt.title('Top-4,May-2020')


plt.subplot(2,2,2)
plt.pie('Jun-20',labels='State',autopct='%1.1f%%', data=df1)
plt.title('Top4,June-2020')

plt.subplot(2,2,3)
plt.pie('May-20',labels='State',autopct='%1.1f%%', data=df2)
plt.title('May-2020')

plt.subplot(2,2,4)
plt.pie('Jun-20',labels='State',autopct='%1.1f%%', data=df2)
plt.title('June-2020')

plt.suptitle('Pie Chart showing the top 4 states and cases of all remaining states as other')
plt.show()

plt.savefig('Plot1.pdf')
```
